### Accuracy resolution every 0.1

```python
groups = pd.cut(df["Pipeline_Acc"], [-0.1] + [i / 10 for i in range(11)])
splits = [d for _, d in df.groupby(groups)]
df0, df1 = splits[0], splits[1]

frames = [df0, df1] 
result = pd.concat(frames)
splits = [result] + splits[2:]
len(splits)
```

```python
for i in range(len(splits)):
    df = splits[i]
    categories = list(df.topic0)
    values = list(df.topic1)
    plt.figure(figsize=(5, 5))
    plt.scatter(categories, values)
    
#     plt.xlim(0 , 1.2)
#     plt.ylim(0 , 1.2)
    plt.xticks(np.arange(0, 1.21, 0.1))
    plt.yticks(np.arange(0, 1.21, 0.1))
    plt.suptitle('Accuracy between '+str(i/10)+' and '+str((i+1)/10)+': topic0-topic1 Plotting')

plt.show()
```
